# KNeighborsClassifier()
To implement k-nearest neighbors vote, sklearn has a function KNeighborsClassifier() which is in the module neighbors.

```python
def knn_comparison(data, k):
    '''
    This function finds k-NN and plots the data.
    '''

    X = data[['x1','x2']].values
    y = data['y'].astype(int).values

    clf = neighbors.KNeighborsClassifier(n_neighbors=k)
    # fit the train set
    clf.fit(X, y)

    # Plotting decision regions
    plot_decision_regions(X, y, clf=clf, legend=2)

    # Adding axes annotations
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Knn with K='+ str(k))
    plt.show()
```

### As the k value increases, the decision surface becomes smooth.

```python
data = pd.read_csv('/content/drive/MyDrive/demo_data/6.overlap.csv', names=['x1', 'x2', 'y'])
for i in [1, 5, 15, 30, 45]:
    knn_comparison(data, i)
```
