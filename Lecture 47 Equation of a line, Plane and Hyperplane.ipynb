```python
def distance(w, x, b=0):
    w_norm = np.linalg.norm(w,2)
    return abs(np.dot(w,x) + b) / w_norm

np.random.seed(100)

w = np.random.rand(6)
b = 1.2
x = np.random.rand(6)

d = distance(w, x, b)
print(d)
```

```python
# how to test that it work?
#let pose
x = [21, 0, 0, 0, 0, 0]
w = [1, 0, 0, 0, 0, 0]
# d should be equal to 21
d = distance(w, x)
d
```
