# Accuracy

Accuracy is one of the criteria for evaluating classification models. Informally, accuracy refers to the proportion of correct predictions made by the model. Formally, accuracy is defined as follows:
#### $$Accuracy = \frac {Number-of-correctly-classified-points}{Total-number-of-points-in-Test-dataset}$$

### Problems of Accuracy as the measure

#### Imbalanced data
Accuracy is only done on the *test data*. Accuracy fails in case of **imbalanced** data.
In case of imbalanced dataset we should never use accuracy as the measure, because a **DUMB**
model can give high accuracy.

```python
Image('Accur_imbalanced.png')
```

#### Probability scores
 
Let's assume we have a test set {x,y} and two models $M_1$, $M_2$ that return a probability scores and predicted values $\hat{y_1}$ $\hat{y_2}$, if we compares the results of model1 and model2 in the table below:

|x|y|$M_1$|$M_2$|$\hat{y_1}$|$\hat{y_2}$|
|--|--|--|--|--|--|
|$x_1$|1|0.9|0.6|1|1|
|$x_2$|1|0.8|0.65|1|1|
|$x_3$|0|0.1|0.45|0|0|
|$x_4$|0|0.15|0.48|0|0|


We can see that the predicted values are exactly the same for the two models, despite the fact that model2 is much less efficient than model1, by looking at the probability scores. We can therefore see that accuracy cannot use probability scores because $\hat{y_1}$ $\hat{y_2}$ says that the two models have the same accuracy which is not true.

# Confusion Matrix

Fisrt let's note that confusion matrix cannot process probability scores i.e: predicted class label are binary (0,1).  
Let assume the test dataset in the table below:

|Datapoint|Actual class label|Predicted class label|
|---------|------------------|---------------------|
|$x_1$|$y_1$|$\hat{y_1}$|
|$x_2$|$y_2$|$\hat{y_2}$|
|$x_3$|$y_3$|$\hat{y_3}$|
|.|.|.|
||.|.|.|
|.|.|.|
|$x_n$|$y_n$|$\hat{y_n}$|

|| 0|1|
|-|-|-|
|0|a|b|
|1|c|d|

a = number of points such that $y_i = 0$ and $\hat {y_i} = 0$  
b = number of points such that $y_i = 1$ and $\hat {y_i} = 0$  
c = number of points such that $y_i = 0$ and $\hat {y_i} = 1$  
d = number of points such that $y_i = 1$ and $\hat {y_i} = 0$  

This matrix is called confusion matrix. If the model is **sensible** `a` and `d` should be high and `b` and `c` should be small.  
The `a`, `b` ,`c` and `d` values have a special names. 

|| 0|1|
|-|-|-|
|0|TN|FN|
|1|FP|TP|

TN: True Negative    
FN: False Negative  
FP: False Positive  
TP: True Positive  

One **trick** to remember this special names:  
let's take the `TN`, if we decompose it, we get two parts. Always the **first** part `T` come from **actual class** and the **second** part `N` come from **predicted class** like image below.



```python
Image('actual.png')
```

We can ask the following questions:

- second part: what is the **predicted** label? (`P`ositive or `N`egative)
- first part: are you correct? (`T`rue or `F`alse)

So for `TN` (0,0) as example the second part represent the predicted label (0)--> `N` negative, the fist part is correct (0)--> `T`.  

Let's take another example: `FP`(0,1)  

what is the predicted label?  
 (predict 1) --> positive `P`  
are you correct?
(actual 0) -->  false`F`

# TPR, FPR, TNR, FNR

Now, the sum of `FN`+ `TP` is called the total number of positives `P` and the sum of `TN` + `FP` is called the total of negatives `N`. And the total number `n` of dataset equal to `n = P + N` <br>

### $TPR = \frac{TP}{P}$  
### $FPR = \frac{FP}{N}$  
### $TNR = \frac{TN}{N}$  
### $FNR = \frac{FN}{P}$  


### Various confusion matrix measurements:
Let's take an example:  
suppose we have an imbalanced test dataset which contains 900 negative points and 100 positive points.
Let's assume we build a model and we got the results below:

```python
Image('model.png')
```

If we look at the four rates, we find:  

TPR = 94%  

TNR = $\frac{850}{900} = 94.44$%  

FPR = $\frac{50}{900} = 5.55$%  

FNR = 6%  

Because TPR and TNR are high and FPR and FNR are small the model is sensible.


Let's take an example of **Dump** model with same data, if the model is dump it will predicted all future query as negative point.
