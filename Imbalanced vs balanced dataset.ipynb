# Imbalanced vs balanced dataset

```python
Image("dataset.png")
```

Imagine we have two class-classification problem, and let's assume we have $D_n$ dataset with $n_1$ positive points and $n_2$ negative points, with $n = n_1 + n_2$. Now if $n_1 \approx n_2$, let assume $n = 1000$, with $n_1 = 580$ and $n_2 = 420$. $n_1$ is approximatly equal to $n_2$. It's called **balanced dataset**, because tere is a good balance between positive and negative points. Now if $n_1$  is significantly less than $n_2$ :  $n_1$  << $n_2$ or $n_2$  << $n_1$, for example if $n_2 = 900$ and $n_1 = 100$, here de disparity is much more, 90% of points in class $n_2$  against 10% in class $n_1$, It's called **imbalanced dataset** .

### How to deal with imbalanced dataset?

Let's take the example of K-NN and suppose that our dataset is severely imbalanced with $n_1 = 50$ (blue points) and $n_2 = 950$ (orange points).

```python
Image('points.png')
```

We can see that the negative class dominates over the positive class. **Now what happens if we have a query point $x_q$?**

```python
Image('queryPoint.png')
```

It is well known that K-NN is based on the nearest neighbors and also on the majority vote, so the majority class will have a big weight, and the result will be biased.

### How to work arround imbalanced dataset issue ?

#### 1- Undersampling:
The first solution is called undersampling.Suppose we have a $D_n$ dataset with 100 positive points and 900 negative points. Undersampling is the fact of creating a new $D'_n$ dataset from the $D_n$ dataset, but which contains all 100 positive points and that we will add **randomly sample** 100 other negative points fromm the 900 negative points.  <br/>

**The sampling must be done randomly**.

```python
Image('undersampling.png')
```

So, we get a new balanced $ D'_n $ dataset, which contains 100 negative points and 100 positive points, and all **the machine learning Modeling will do in**  $ D'_n $ instead of  $ D_n $.

```python
Image('balancedDataset.png')
```

With **undersampling**, there is one problem insofar as there is **data loss** in case of under sampling. We can overcome this situation by **Oversampling** the data.

 #### 2- Oversampling

Oversampling is taking the same data point **repeatedly**.  
Random oversampling involves randomly duplicating examples from the minority class and adding them to the training dataset.  <br/>

Examples from the training dataset are selected **randomly with replacement**. This means that examples from the **minority** class can be chosen and added to the new *more balanced* training dataset **multiple** times; they are selected from the original training dataset, added to the new training dataset, and then returned or replaced in the original dataset, allowing them to be selected again.
