# Precision, Recall an F1-score

```python
Image('pr.png')
```

Precision and recall are often used in **Information retreival** problems like search engines.  

### $Precision = \frac{TP}{TP + FP}$  

- Precision means that of all the points of the model **predicted** to be positif, what percentage % of them are **actually** positive.
- Precision only computes the positive class predicted rate.

Recall is very simple, is not but the true positive rate `TPR`. The name changed because they came from different area, in statistics it's called `TPR`, in info retreival it's called `Recall`  
### $Recall = TPR = \frac{TP}{P}$  

- Recall means of all the points **actually** positive, how many of them are **predicted** positive.  

F1-score, it's the idea to combine the precision and recall in one measure.  
Always we want that precision and recall be high.  

### $F1-score = 2 \times \frac{precision \times recall}{precision + recall}$  

- F1-score is the harmonic mean of the precision and recall.
- Precision and recall are more interpretable, than F1-score.
