# The dot product

```python
## many ways to compute the dot product

np.random.seed(100)
v1 = np.random.rand(2)
v2 = np.random.rand(2)

# method 1
dp = sum( np.multiply(v1,v2) )

# method 2
dp = np.dot( v1,v2 )

# method 3
dp = np.matmul( v1,v2 )
```

```python
def dot_product(v1, v2):
    return np.dot(v1, v2)
```

```python
w = dot_product(v1, v2)
print(w)
```

```python
X = np.zeros(2)
x,y = zip(X, v1)
plt.plot(x, y, label='v1')
x,y = zip(X, v2)
plt.plot(x,y, label='v2')
plt.legend()
plt.show()
```

### Angle between two vectors

```python
Image(url= "https://i.imgur.com/zvFu6Du.png", width=400)
```

```python
def angle(u,v, degree=False):
    u_norm = np.linalg.norm(u, 2)
    v_norm = np.linalg.norm(v, 2)
    
    if degree:
        return np.degrees(math.acos(np.dot(u,v) * 1/u_norm * 1/v_norm))
    else:
        return math.acos(np.dot(u,v) * 1/u_norm * 1/v_norm)
```

```python
np.random.seed(10)
u = np.random.rand(2)
v = np.random.rand(2)
theta = angle(u,v)
print(theta)
```

```python
v1 = np.array([2, 4, -3])
v2 = np.array([0,-3, -3])
theta = angle(v1,v2)
print(round(theta, 3))
```

```python
X = np.zeros(3)
x,y,z = zip(X, v1)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(x,y,z, label='v1')

x,y,z = zip(X, v2)
ax.plot(x,y,z, label='v2')

plt.legend()
plt.show()
```
