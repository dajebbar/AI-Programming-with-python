We will illustrate the slices on the following string, however remember that this mechanism works with all the sequences that we will see later, such as lists or tuples.

```python
my_str = "abcdefghijklmnopqrstuvwxyz"
print(my_str)
```

### Slice without step

a slice is used to designate a whole range of elements in a sequence. So we can write:

```python
my_str[2:6]
```

### Start and end conventions

Beginners sometimes have trouble with the terminals. It should be remembered that:
- the indices ** start ** as always ** at zero **;
- the first index at the start is ** included **;
- the second fine index is ** excluded **;
- we get at all end-start items in the result.
<br>
<p>So, above, the result contains 6 - 2 = 4 elements.</p>

To help you remember the start and end conventions, remember that you want to be able to easily juxtapose two slices that have a common bound.

That is to say with:

```python
Image('Img/slice.png')
```

```python
# my_str[a:b] + my_str[b:c] == my_str[a:c]
my_str[0:3] + my_str[3:7] == my_str[0:7]
```

### Omitted terminals

We can omit a bound:

```python
# if we omit the first terminal, it means that
# the slice starts at the start of the object
my_str[:6]
```

```python
# and of course it's the same if we omit the second terminal
my_str[24:]
```

```python
# or even omit both bounds, in which case we
# make a copy of the object - more on this later
my_str[:]
```

### Negative indices

We can use negative indices to count from the end:

```python
my_str[3:-3]
```

```python
my_str[-3:]
```

## Slice with step

It is also possible to specify a step, so as to choose for example, in the given range, only one element out of two:

```python
# the step is specified after a second colon (:)
# here we will choose every second character in the range [3: -3]
my_str[3:-3:2]
```

As you can guess, the third element of the slice, here 2, determines the pitch. We therefore retain, in the chain defghi ... only d, then f, and so on.

We can therefore specify the end bound (here -3) with a little freedom, since here we would obtain an identical result with -4.

```python
my_str[3:-4:2]
```

### Negative step

It is even possible to specify a negative pitch. In this case, in a slightly counter-intuitive way, you have to specify a beginning (the first index of the slice) which is more to the right than the end (the second index).

To take an example, as the element of index -3, that is to say x, is more to the right than the element of index 3, that is to say d, obviously if we do not not specify the step (which amounts to choosing a step equal to 1), we would obtain an empty list:

```python
my_str[-3:3]
```

If now we specify a negative step, we obtain this time:

```python
my_str[-3:3:-2]
```
