### Reading data from excel file

```python
df = pd.read_excel('data/sales-data.xlsx',sheet_name='Store Sales')
df.head()
```

#### How to deal with the case where the column is considered as a row? As this example shows

```python
df = pd.read_excel('data/sales-data.xlsx',sheet_name='Dictionary')
df
```

#### To fix it we use `skiprows` and `uscols`

```python
df = pd.read_excel('data/sales-data.xlsx',sheet_name='Dictionary',
                    skiprows=1, usecols=['Variable', 'Description'])
df
```

#### How to deal with the case without a header?

```python
df = pd.read_table('data/Emp-info.txt', sep=',')
df.head(3)
```

#### We use the `header` and `names` attributes.

```python
df = pd.read_table('data/Emp-info.txt', sep=',', 
                    header=None, 
                    names=['Name', 'Gender', 'Post', 'Dept', 'Countries'])
df
```
