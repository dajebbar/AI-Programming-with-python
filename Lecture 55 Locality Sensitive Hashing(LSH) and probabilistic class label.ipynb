# Hashing vs Locality Sensitive Hashing

LSH refers to a family of functions (known as LSH families) to hash data points into buckets so that data points near each other are located in the same buckets with high probability, while data points far from each other are likely to be in different buckets. This makes it easier to identify observations with various degrees of similarity.  
Although these algorithms don’t guarantee to give you the exact answer, more often than not they’ll be provide a good approximation. These algorithms are faster and scalable.

# Locality Sensitive Hashing for Cosine Similarity

### Implementation steps for LSH
The picture summarizes the generation of Hyperplanes, note the points below about this hyperplane generation math.  
**How to generate the random plane $\pi$ ?**  
- As this $\pi$ hyperplane is passing through origin its equation is $\omega^T . x = 0$ And if Equation of a plane is $\omega^T . x = 0$ that means $\omega$ is nothing but the normal to the plane by definition of Normal of a Vector.
- Random hyperplane is generated using $\omega^T . x = 0$ where $\omega$ is "normal" to the plane passing through origin, since $\omega$ is normal to the plane we have taken values from N(0,1) in order to initialize its value.
